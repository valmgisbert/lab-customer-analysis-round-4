```python
# X-y split
# drop the ones with neg corr to target
X = data.drop(['income', 'number_of_open_complaints', 'total_claim_amount'], axis = 1)
y = data['total_claim_amount']
```

```python
X_num = X.select_dtypes(include = np.number) #get all numerical columns
X_num
```

```python
# Normalize (numerical)
transformer = MinMaxScaler().fit(X_num) 
X_norm = transformer.transform(X_num) #output always an array

X_normalized = pd.DataFrame(X_norm, columns=X_num.columns)
X_normalized
```

### One/Hot label encoding

```python
X_cat = X.select_dtypes(include = object)
X_cat
```

```python
X_cat.columns.values.tolist() #eval how many categories we have
# Customer cat doesn't make sense in labeling as it's an unique value, so we drop it
# Effective To Date cat doesn't make sense to keep, format is weird to onehot
X_cat = X_cat.drop(['customer', 'effective_to_date'], axis=1)
X_cat
```

```python
# we can onehotencode directly to the dataframe
encoder = OneHotEncoder().fit(X_cat)

cat_encoded = encoder.transform(X_cat).toarray()
cat_onehot_encoded = pd.DataFrame(cat_encoded, columns=encoder.get_feature_names_out()) #to get the names of the feats

cat_onehot_encoded
```

### Concat dataframes

```python
X = pd.concat([X_normalized, cat_onehot_encoded], axis=1)
X.head()
```

## Linear regression

### Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape) #80% of it
print(X_test.shape) #20% of it
print(y_train.shape)
print(y_test.shape)
```

### Apply linear regression

```python
lm = linear_model.LinearRegression()
lm.fit(X_train, y_train) #model training
```

```python
predictions_test = lm.predict(X_test) #model prediction
```

## Model Validation

#### R2

```python
#R2 - test
r2_score(y_test, predictions_test)
#0.784 with trimming some of the tails in dataset - check Lilit's lab
```

#### MSE

```python
#MSE - train ones are usually a little higher 
mse = mean_squared_error(y_test, predictions_test) 
mse

#19 before reducing
#17103.606 after reducing
```

#### RMSE

```python
#RMSE
rmse = np.sqrt(mean_squared_error(y_test, predictions_test)) 
rmse
```

```python
#MAE
mae = mean_absolute_error(y_test, predictions_test)
mae
```
